```python
import json
import pandas as pd
from pandas import DataFrame,Series
```

```python
with open('social-pulse-trentino.geojson','r') as f:
    file = json.load(f)
file
```

```python
fram = DataFrame(file['features'])
fram
```

```python
frcity = fram.groupby('municipality.name').sum()
city = frcity.index
city
```

```python
timestamp = [1383260400]
for i in range(1,8784):
    timestamp.append(600+timestamp[i-1])
len(timestamp)
```

```python
def get_city_tw_number_series(cityname):
    twi_number = pd.DataFrame(index=timestamp,columns=[cityname])
    for i in range(len(timestamp)):
        if i == len(timestamp)-1:
            tw = fram.loc[(fram['municipality.name'] == cityname) & (fram['timestamp'] < 1388530800) & 
                (fram['timestamp'] >= 1388530200)]
            number = tw.count()
            number = number['timestamp']
            twi_number.loc[timestamp[i]] = number
            break
        else:
            tw = fram.loc[(fram['municipality.name'] == cityname) & (fram['timestamp'] < timestamp[i+1]) & 
                (fram['timestamp'] >= timestamp[i])] #1383260400 === > 1383346200  步长 600
            number = tw.count()
            number = number['timestamp']
            twi_number.loc[timestamp[i]] = number
    twi_number['timestamp'] = twi_number.index
    return twi_number
```

```python
def get_wholecity_twi_data():
    for i in range(len(city)):
        t = get_city_tw_number_series(city[i])
        if i == 0:
            whole_data = t
        else:
            whole_data = pd.DataFrame.merge(whole_data,t,how='outer',on='timestamp')
    return whole_data
```

```python
whole = get_wholecity_twi_data()
```
