```python
file = pd.read_csv('trn_twi_process.csv',encoding='gb18030')
file
```

```python
index = file.T.index
```

```python
index
```

```python
max = 0
sumn = 0
for i in range(1,len(index)):
    if i == 2:
        continue
    number = file[index[i]].sum()
    print(number)
    if number > max:
        max = number
        maxindex = i
    sumn = sumn + number
```

```python
print(max,maxindex,sumn)
```
