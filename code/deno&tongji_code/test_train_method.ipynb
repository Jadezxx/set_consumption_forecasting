```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import LSTM 
from keras.layers import Bidirectional
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets('mnist/', one_hot=True) 
import time

        
batch_size = 100
timesteps = 28 
data_dim = 28 
num_classes = 10

time_start1 = time.time()

model = Sequential()
model.add(Bidirectional(LSTM(64,return_sequences=True,dropout=0.2, recurrent_dropout=0.2),input_shape=(timesteps, data_dim)))
model.add(LSTM(128,return_sequences=False))
model.add(Dense(10, activation='relu'))
model.add(Activation('softmax'))
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

time_media1 = time.time()

print('Train...')
def train_network(): 
    step = 0
    test_x = []
    test_y = []
    train_x = []
    train_y = []
    for i in range(100):

        x_test = mnist.test.images[i*100:i*100+100,:]
        y_test = mnist.test.labels[i*100:i*100+100,:]  
        test_x.extend(x_test)
        test_y.extend(y_test)
            
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        train_x.extend(batch_x)
        train_y.extend(batch_y)
        
    x_train = np.array(train_x).reshape(10000,timesteps,data_dim)
    test_x = np.array(test_x).reshape(10000,timesteps,data_dim)
    train_y = np.array(train_y).reshape(10000,10)
    test_y = np.array(test_y).reshape(10000,10)
    #print(train_y)
    # 
    model.fit(x_train, train_y,batch_size=batch_size,epochs=15,validation_data=(test_x, test_y))
    return model
train_network()
score, acc = model.evaluate(test_x, test_y,batch_size=batch_size)
time_end1 = time.time()
```

```python
a = time_media1 - time_start1
b = time_end1 - time_media1
c = time_end1 - time_start1
print(a,b,c)
```
