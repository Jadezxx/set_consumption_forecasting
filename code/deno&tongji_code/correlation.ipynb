## 定义各个文件的路径

```python
filename_set_nov = 'C:/program2/tensorflow/data/set_consumption_nov.csv' # 0:00 
filename_set_dec = 'C:/program2/tensorflow/data/set_consumption_dec.csv' # 0:00 
filename_traffic_nov = 'C:/program2/tensorflow/data/traffic_nov.csv'# 7:00
filename_traffic_dec = 'C:/program2/tensorflow/data/traffic_dec.csv'# 7:00
filename_tem_nov = 'C:/program2/tensorflow/data/temperature_nov_resample.csv'# 0:00  36个id
filename_tem_dec = 'C:/program2/tensorflow/data/temperature_dec_resample.csv'# 0:00  36个id
filename_sms_call_dec = 'C:/program2/tensorflow/data/sms_call_dec.csv'# 7:00
filename_sms_call_nov = 'C:/program2/tensorflow/data/sms_call_nov.csv'# 7:00
filename_traffic_de_noising_nov = 'C:/program2/tensorflow/data/traffic_de_noising_nov.csv'# 7:00
filename_traffic_correlation_nov = 'C:/program2/tensorflow/data/traffic_correlation_nov.csv'# 7:00
filename_traffic_corr_id = 'C:/program2/tensorflow/data/traffic_corr_id.csv'
```

## 获得电力数据

```python
set_nov = pd.read_csv(filename_set_nov)#'10867', '10868', '10980', '10981', '10982', '10983', '10984', '11097'
set_nov = set_nov.drop('Unnamed: 0',1)
set_nov = set_nov.drop('Unnamed: 0.1',1)
set_nov = set_nov.drop('Unnamed: 0.1.1',1)
set_index = set_nov.columns
```

## 获得流量数据

```python
traffic_nov = pd.read_csv(filename_traffic_nov)#'id_9991', 'id_9992', 'id_9993', 'id_9994', 'id_9995', 'id_9996'
traffic_nov = traffic_nov.drop('Unnamed: 0',1)
traffic_nov = traffic_nov.drop('Unnamed: 0.1',1)
traffic_nov = traffic_nov.drop('time',1)
traffic_index = traffic_nov.columns
```

## 获得去噪流量数据

```python
traffic_de_noising_nov = pd.read_csv(filename_traffic_de_noising_nov)#'9991', '9992', '9993', '9994', '9995', '9996'
traffic_de_noising_nov = traffic_de_noising_nov.drop('Unnamed: 0',1)
traffic_de_noising_index = traffic_de_noising_nov.columns
```

## 获得电力与去噪流量的相关系数

```python
# -*- coding: UTF-8 -*-
import numpy as np
corr_traffic = []

for j in range(len(set_index)):
    corr1 = []
    if str(set_index[j]) not in traffic_de_noising_index:   
        continue
    y1 = set_nov[str(set_index[j])]
    y2 = traffic_de_noising_nov[str(set_index[j])]
    correlation = y1.corr(y2, method = 'pearson')
    corr1.append(set_index[j])
    corr1.append(correlation)
    corr_traffic.append(corr1)
corr_traffic
```

## 获得电力与流量数据的相关系数

```python
# -*- coding: UTF-8 -*-
import numpy as np
corr_traffic = []

for j in range(len(set_index)):
    corr1 = []
    if 'id_'+str(set_index[j]) not in traffic_index:   
        continue
    y1 = set_nov[str(set_index[j])]
    y2 = traffic_nov['id_'+str(set_index[j])]
    correlation = y1.corr(y2, method = 'pearson')
    corr1.append(set_index[j])
    corr1.append(correlation)
    corr_traffic.append(corr1)
#print(len(corr_traffic))
corr_traffic
```

## 计算流量数据的相关系数

```python
fu_corr = []
no_corr = []
low_corr = []
media_corr = []
high_corr = []
data = corr_traffic
for i in range(len(data)):
    if data[i][1] > 0.8:
        high_corr.append(corr_traffic[i][0])
    elif data[i][1] > 0.5 and data[i][1] <= 0.8:
        media_corr.append(corr_traffic[i][0])
    elif data[i][1] > 0.3 and data[i][1] <= 0.5:
        low_corr.append(corr_traffic[i][0])
    elif data[i][1] > 0 and data[i][1] <= 0.3:
        no_corr.append(corr_traffic[i][0])
    elif data[i][1] < 0:
        fu_corr.append(corr_traffic[i][0])
print('共有：',len(fu_corr),'个负相关地区，分别为：')
print(fu_corr[:10],'等...')
print('共有：',len(no_corr),'个不存在相关地区，分别为：')
print(no_corr[:10],'等...')
print('共有：',len(low_corr),'个低度相关地区，分别为：')
print(low_corr[:10],'等...')
print('共有：',len(media_corr),'个显著相关地区，分别为：')
print(media_corr[:10],'等...')
print('共有：',len(high_corr),'个高度相关地区，分别为：')
print(high_corr[:10],'等...')
```
