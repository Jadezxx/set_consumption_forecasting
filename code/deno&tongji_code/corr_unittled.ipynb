```python
from pylab import *
import numpy as np
from scipy import signal

nSampleNum = 5120 
ncount = 2048.0
df = nSampleNum / ncount
sampleTime = ncount / nSampleNum
freqLine = 800

x = [i for i in range(4320)]#时域波形x轴坐标
y = traffic_nov['id_2737']

fft = np.fft.fft(y)[0:4320]  #调用fft变换算法计算频域波形
fftx = np.linspace(0,500,4320)  #频域波形x轴坐标311)
fig = figure(figsize=[50,20])
ax1 = fig.add_subplot(4,1,3)
ax1.plot(x,y)
xlabel('time(s)')
ylabel('amplitude')
title('time domain graph')

print(fft.shape)

ax2 = fig.add_subplot(4,1,1)
ax2.plot(fftx,abs(fft))
xlabel('freqency(Hz)')
ylabel('amplitude')
title('frequency domain graph')

fft[90:] = 0 #高通滤波

ax4 = fig.add_subplot(4,1,2)
ax4.plot(fftx,abs(fft))
xlabel('freqency(Hz)')
ylabel('amplitude')
title('frequency domain graph')

f1shift = np.fft.ifft(fft)
ax3 = fig.add_subplot(4,1,4)
ax3.plot(x,f1shift)

correlation1 = set_nov['2737'].corr(pd.Series(f1shift), method = 'pearson')#((0.29395913590865563+0.1381417427428661j)
print(correlation1)

correlation2 = set_nov['2737'].corr(y, method = 'pearson')#((0.29395913590865563+0.1381417427428661j)
print(correlation2)

ax5 = ax3.twinx()
ax5.plot(x,set_nov['2737'],'r')

show()
```
