```python
import pandas as pd
import numpy
import matplotlib.pyplot as plt
```

画降水分布图

```python
#########          加载一天的温度或降水数据              #############

def get_day_file(filename):                
    day_file = pd.read_csv(filename)
    return day_file
```

```python
for i in range(1,8):
    filename = 'per_2013-11-'+str(i)+'.csv'
    data = get_day_file(filename)
    if i == 1:
        y = list(data['1470'])
    else:
        y.extend(data['1470'])
len(y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = [i for i in range(len(y))]
fig = plt.figure(figsize=(25,10))
ax = fig.add_subplot(1,1,1)
ax.set_xticks(np.linspace(0,672,8))
ax.set_xticklabels(('11-1-0:0','11-2-0:0','11-3-0:0','11-4-0:0',
                    '11-5-0:0','11-6-0:0','11-7-0:0','11-8-0:0'))
ax.plot(x,y)
plt.savefig('per_11-1_week.jpg')
plt.show()
```

画出一个点一天的图：

```python
#########          画出某一个id一天的降水量              #############
def get_per_id_day(square_id,day_filename):
    day_file = get_day_file(day_filename)
    x = day_file.index
    y = day_file[str(square_id)]
    fig = plt.figure(figsize=(15,10))
    ax = fig.add_subplot(1,1,1)
    ax.plot(x,y)
    plt.title(str(square_id)+'_per_day')
    plt.xlabel('time')
    plt.ylabel('percipitation')
    ax.set_xticklabels(['start','00:00','05:00','10:00','15:00','20:00','finish'])
    plt.show()
```
