```python
def get_id_consum_proption():
    for i in range(len(line_id)):
        id_consum = pd.DataFrame(columns=['id',str(line_id[i])])
        f_groupline = line.loc[line['line'] == line_id[i]]
        total_equip = f_groupline['equipment'].sum()
        #square_id = list(f_groupline['square_id'])
        f_line = f_groupline.set_index('square_id') #或者（）
        id_consum['id'] = f_groupline['square_id']
        id_consum = id_consum.set_index('id')
        f_line['equipment'] = f_line['equipment']/total_equip
        id_consum[str(line_id[i])] = f_line['equipment']
        id_consum['id'] = id_consum.index
        if i == 0:
            id_consum_merge = id_consum
        if i != 0:
            id_consum_merge = pd.merge(id_consum_merge,id_consum,how='outer',on='id')
    return id_consum_merge
```

```python
def time2stamp(time):
    import time
    timeArray = time.strptime(time, "%Y-%m-%d %H:%M")
    #转换成时间戳
    timestamp = time.mktime(timeArray)
    return timestamp
```

```python
def stamp2time(stamp):
    import time
    #转换成localtime
    time_local = time.localtime(stamp)
    #转换成新的时间格式(2016-05-05 20:28:54)
    dt = time.strftime("%Y-%m-%d %H:%M",time_local)
    return dt
```

```python
def get_timepoit_line_consumption(timestp):
    #n = 0
    id_consum_1 = id_consum_merge
    id_consum_1 = id_consum_1.fillna(0)
    for i in range(len(line_id)):
        cons = set_consumption.loc[(set_consumption['time']==stamp2time(timestp)) & (set_consumption['line']==line_id[i])]
        if len(cons) != 0:
            #print(i)#cons = set_partion.loc[set_partion['line'] == line_name[i]]['consumption']  
            id_consum_1[line_id[i]] = id_consum_1[line_id[i]]*float(cons['consumption'])
            #n+=1
        else:
            id_consum_1[line_id[i]] = id_consum_1[line_id[i]]*float(0)
            #n+=1
            #print(i)
    #print(n)
    return id_consum_1
```
